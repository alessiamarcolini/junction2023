```python
import os


import pandas as pd
```

```python
input_folder = "../data/"
```

```python
df = pd.read_csv(os.path.join(input_folder, "eletricity_price_index.csv"))
```

```python
df = df.loc[:, ~df.columns.str.contains('Unnamed')]
```

```python
df
```

```python
df = df[df.iloc[:, 1] != ':']
df = df.set_index('TIME').T
```

```python
df.to_csv(os.path.join(input_folder, "processed_seletricity_price_index.csv"), index_label="time")
```
